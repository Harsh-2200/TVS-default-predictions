## Importing Necessary Libraries

```python
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score
```

## 1 Data Cleaning

```python
data = pd.read_csv('Train_Set_TVS.csv')
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['Compensation']    = data['Compensation'].fillna(data['Compensation'].mean())
data['Annuity_amt']     = data['Annuity_amt'].fillna(data['Annuity_amt'].mean())
data['Experience']      = data['Experience'].fillna(data['Experience'].mean())
data['Company_Rating']  = data['Company_Rating'].fillna(data['Company_Rating'].mean())
data['c1_rated']        = data['c1_rated'].fillna(data['c1_rated'].mean())
data['c2_rated']        = data['c2_rated'].fillna(data['c2_rated'].mean())
```

```python
data.dropna(axis = 0 , inplace = True)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
X = data.iloc[:,[1,2,3,4,6,7,8,9,10,11,12,13,15,16,17,19,20,21,22,23,24,26,27,28,29,31]].values
Y = data.iloc[:,-1].values
```

```python
cat = [1,4,5,6,7,8,9,10,11,12,14,15,16,19,21,22]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), cat)], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

## 3 Model Building

In this you can build various ML models and try tuning the parameters involved to achieve better score

```python
test = pd.read_csv('Test_Set_TVS.csv').values
X_test = data.iloc[:,[1,2,3,4,6,7,8,9,10,11,12,13,15,16,17,19,20,21,22,23,24,26,27,28,29,31]].values
Y_test = data.iloc[:,-1].values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), cat)], remainder='passthrough')
X_test = np.array(ct.fit_transform(X_test))
```

```python
X_test  = sc.transform(X_test)
```

## Model 1

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression(random_state = 0)
logistic.fit(X, Y)

```

```python
y_pred_log = logistic.predict(X_test)
print(np.concatenate((y_pred_log.reshape(len(y_pred_log),1), Y_test.reshape(len(Y_test),1)),1))
```

```python
cm = confusion_matrix(Y_test, y_pred_log)
print(cm)
accuracy_score(Y_test, y_pred_log)*100
```

## Model 2 

```python
from sklearn.tree import DecisionTreeClassifier
Decision = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
Decision.fit(X, Y)

```

```python
y_pred_dec = Decision.predict(X_test)
print(np.concatenate((y_pred_dec.reshape(len(y_pred_dec),1), Y_test.reshape(len(Y_test),1)),1))
```

```python
cm = confusion_matrix(Y_test, y_pred_dec)
print(cm)
accuracy_score(Y_test, y_pred_dec)*100
```

## Model 3

```python
from sklearn.ensemble import RandomForestClassifier
Random = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
Random.fit(X, Y)

```

```python
y_pred_ran = Random.predict(X_test)
print(np.concatenate((y_pred_ran.reshape(len(y_pred_ran),1), Y_test.reshape(len(Y_test),1)),1))
```

```python
cm = confusion_matrix(Y_test, y_pred_ran)
print(cm)
accuracy_score(Y_test, y_pred_ran)*100
```
